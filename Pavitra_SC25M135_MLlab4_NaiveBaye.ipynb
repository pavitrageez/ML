```python
#Q1 Perform Naive Bayes classification on IRIS dataset. Train a Naive Bayes classifier and evaluate model's accuracy.
```

```python
import pandas as pd
import numpy as np
iris = pd.read_csv('/content/Iris.csv')
print(iris.head())
print(iris.shape)
print(iris.columns)
```

```python
#Load the X,Y data and target features
X = iris.iloc[:,1:5]
Y = iris.iloc[:,5]
print(X.head())
print(Y.head())

iris['Species_name'].unique()
```

```python
# Merge feature names with X and target names for Y
feature_names = X.columns
target_names = iris['Species_name'].unique()
print(feature_names)
print(target_names)
```

```python
#Train Split and evaluate
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print('X_train :')
print(X_train.head())
print('X_test :')
print(X_test.head())
print('Y_train :')
print(Y_train.head())
print('Y_test :')
print(Y_test.head())
```

```python
#Import naive bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

#Train the model in Naive bayes
model = GaussianNB()
model.fit(X_train, Y_train)
```

```python
#Predict the model's accuracy in percentage
from sklearn.metrics import accuracy_score
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
accuracy_percentage = accuracy * 100
print('Accuracy percentage:', accuracy_percentage)
```

```python
#For  petal_width=2.2, petal_length=6.5. sepal_width=3.5, sepal_length=6.5,calculate the probability that sample belongs to Setosa? If not, specify the class where it belongs from the scores of probabilities
```

```python
sample = np.array([[2.2, 6.5, 3.5, 6.5]])
probability = model.predict_proba(sample)
print('Probability of sample belonging to each class:')
for name, p in zip(target_names, probability[0]):
    print(f'{name}: {p*100:.2f}%')
print("Predicted class:", model.predict(sample))
```

```python
#If sample does not belong to Setosa, print the correct class based on score
if probability[0][0] < probability[0][1]:
    print("Predicted class:", target_names[1])
    print("Score:", probability[0][1])
else:
    print("Predicted class:", target_names[0])
```

**Conclusion:** From the given sample values, the predicted class is found to be Versicolor with a 100% accuracy compared to Setosa.

**Naive Bayes classification on Basketball dataset**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

basketball = pd.read_csv("/content/Basketball.csv")
print(basketball.head(14))
print(basketball.columns)
```

```python
#Encode categorical data
encoders = {}

for col in basketball.columns:
    le = LabelEncoder()
    basketball[col] = le.fit_transform(basketball[col])
    encoders[col] = le
    print(f"{col} : {le.classes_}")
```

```python
#Train-Split and evaluate
from sklearn.model_selection import train_test_split
X = basketball.drop('Play', axis=1)
Y = basketball['Play']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print('X_train :')
print(X_train.head())
print('X_test :')
print(X_test.head())
print('Y_train :')
print(Y_train.head())
print('Y_test :')
print(Y_test.head())
```

```python
# Apply Naive Baye
from sklearn.naive_bayes import GaussianNB

# Train a new Naive Bayes model for the basketball dataset
basketball_model = GaussianNB()
basketball_model.fit(X_train, Y_train)
```

```python
# Predict the 3 given cases of Outlook: 1) Sunny, Hot, High, Weak;2) Overcast, Mild, High, Strong; 3) Sunny, Cool, Normal, Weak
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Loading the original basketball dataset again to correctly fit the encoders
original_basketball_df = pd.read_csv("/content/Basketball.csv")

# Re-create and fit the encoders correctly
# This ensures the encoders are fitted on string data, not already encoded numerical data.
new_encoders = {}
for col in original_basketball_df.columns:
    le = LabelEncoder()
    le.fit(original_basketball_df[col]) # Fit on original string data
    new_encoders[col] = le

# Now use new_encoders for prediction
cases = [
    ["Sunny","Hot","High","Weak"],
    ["Overcast","Mild","High","Strong"],
    ["Sunny","Cool","Normal","Weak"]
]

for i, c in enumerate(cases, 1):
    c_enc = [
        new_encoders["Outlook"].transform([c[0]])[0],
        new_encoders["Temperature"].transform([c[1]])[0],
        new_encoders["Humidity"].transform([c[2]])[0],
        new_encoders["Wind"].transform([c[3]])[0]
    ]

    # Use the basketball_model trained specifically for this dataset
    probs = basketball_model.predict_proba([c_enc])[0]
    pred = basketball_model.predict([c_enc])[0]

    print(f"\nCase {i}: {c}")
    print("P(No)  =", probs[new_encoders["Play"].transform(["No"])][0]) # Use encoder to get index for 'No'
    print("P(Yes) =", probs[new_encoders["Play"].transform(["Yes"])][0]) # Use encoder to get index for 'Yes'
    print("Prediction:", new_encoders["Play"].inverse_transform([pred])[0])
```

Conclusion:
The Naive Bayes classifier was trained on the basketball dataset using the features Outlook, Temperature, Humidity, and Wind.
For the given test cases, the predicted results are:
Case 1 (Sunny, Hot, High, Weak) - No
Case 2 (Overcast, Mild, High, Strong) - Yes
Case 3 (Sunny, Cool, Normal, Weak) - Yes.
Hence, basketball play is predicted to be possible in Cases 2 and 3, and not possible in Case 1.
